```python
essentialherodata = dog[['account_id','hero_id', 'gold_per_min', 'xp_per_min', 'kills','tower_damage']]
superhero= pd.read_csv('Dota2_DataSet/hero_names.csv')
hero = dict(zip(superhero['hero_id'],superhero['localized_name']))
hero[0] = 'Unknown'
essentialherodata['hero'] = essentialherodata['hero_id'].apply(lambda id : hero[id])
name = 'hero'
col = essentialherodata.pop(name)
essentialherodata.insert(0, name, col)
essentialherodata.head(200)
```

```python
meanherodata = essentialherodata.groupby('hero').mean()
meanherodata.drop('Unknown',inplace=True)
meanherodata.head(200)
```

```python
from sklearn.cluster import KMeans
cluster2 = meanherodata
distortions2 = []
inertias2 = []
mapping10 = {}
mapping20 = {}
K2 = range(1, 10)
for k in K2: 
    kmeanModel = KMeans(n_clusters=k).fit(cluster2)
    kmeanModel.fit(cluster2)
    distortions2.append(sum(np.min(cdist(cluster2, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / cluster2.shape[0])
    inertias2.append(kmeanModel.inertia_)
    mapping10[k] = sum(np.min(cdist(cluster2, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / cluster2.shape[0]
    mapping20[k] = kmeanModel.inertia_
```

```python
for key, val in mapping10.items():
    print(f'{key} : {val}')
```

```python
for key, val in mapping20.items():
    print(f'{key} : {val}')
```

```python
kmeans2 = KMeans(init="random", n_clusters=4,n_init=10,max_iter=3000, random_state=102)
kmeans2.fit(cluster2)
```

```python
cluster2['kmeans2'] = kmeans2.labels_
kmeans2_stats = cluster2.groupby(['kmeans2']).mean()
kmeans2_statmeans = kmeans2_stats.mean(axis=0)
kmeans2_range = kmeans2_stats.max(axis=0) - kmeans2_stats.min(axis=0)
kmeans2_norm = (kmeans2_stats - kmeans2_statmeans) / kmeans2_range
```

```python
fig, (axis1, axis2) = plt.subplots(2,1,figsize=(14,14))
kmeans2_norm.iloc[:,:7].plot.bar(ax=axis2).legend()
```

```python
cluster2.loc[(cluster2['kmeans2']==0),'hero_role'] = 'Soft-carry'
cluster2.loc[(cluster2['kmeans2']==1),'hero_role'] = 'Support'
cluster2.loc[(cluster2['kmeans2']==2),'hero_role'] = 'Offlaner/All-rounder'
cluster2.loc[(cluster2['kmeans2']==3),'hero_role'] = 'Hard-Carry'
```
