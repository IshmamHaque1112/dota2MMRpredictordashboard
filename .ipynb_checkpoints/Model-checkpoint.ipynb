```python
# Import our libraries 

# Pandas and numpy for data wrangling
import pandas as pd
import numpy as np

# Seaborn / matplotlib for visualization 
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline

# Import the trees from sklearn
from sklearn import tree

# Helper function to split our data
from sklearn.model_selection import train_test_split

# Helper fuctions to evaluate our model.
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, roc_auc_score 

# Helper function for hyper-parameter turning.
from sklearn.model_selection import GridSearchCV

# Import our Decision Tree
from sklearn.tree import DecisionTreeClassifier 

# Import our Random Forest 
from sklearn.ensemble import RandomForestClassifier
```

```python
#df = pd.read_csv('Dota2_dataset/players.csv')
df = pd.read_csv('data/players.csv')
df.head()
```

```python
df_player_rating = pd.read_csv('data/player_ratings.csv')
df_player_rating.head()
```

```python
#remove anonymous players
select_features = df['account_id'] != 0

df = df[select_features]
#display all of our rows
pd.set_option("display.max_columns", None)
df.head()

select_condition_3 = df_player_rating['account_id'] > 0
df_player_rating = df_player_rating[select_condition_3]
df_player_rating.tail()
```

```python
df_p = pd.merge(df_player_rating, df, left_on='account_id', right_on='account_id')

def int_to_float(x):
    return int(x)

df_p['trueskill_rating'] = df_p['trueskill_mu'].apply(int_to_float)
df_p.iloc[10:20]
```

```python
#engineer our selected features in order to predict gold earned per minute
selected_features = ['gold_spent', 'gold', 'xp_per_min', 'kills', 'deaths', 'assists', 'denies', 'gold_per_min', 'level', 'total_wins', 'hero_damage', 'hero_healing', 'tower_damage' , 'item_0', 'item_1', 'item_3', 'item_4', 'item_5']

X = df_p[selected_features]

y = df_p['trueskill_rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
pd.set_option("display.max_rows", None)
df.isnull().sum()
```

```python
model = RandomForestClassifier(max_depth=2)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
print("Accuracy Score: %f" % accuracy)

precision = precision_score(y_true=y_test, y_pred=y_pred, average='micro')
print("Precision Score: %f" % precision)

recall = recall_score(y_true=y_test, y_pred=y_pred, average='micro')
print("Recall Score: %f" % recall)

f1 = f1_score(y_true=y_test, y_pred=y_pred, average='micro')
print('F1 Score: %f' % f1)
```
